```python
from time import sleep
from selenium import webdriver
# this is tested on Firefox or you can use "webdriver.Chrome()"
browser = webdriver.Firefox()
browser.get('https://www.facebook.com/marketplace/108164422544464/vehicles?maxPrice=10000&maxMileage=100000&minYear=2010&exact=false')


```

```python
from selenium.webdriver.common.by import By
from IPython.display import display,HTML
```

```python
# https://www.facebook.com/marketplace/108164422544464/cars?minPrice=1000&maxPrice=7000&exact=false

elements = browser.find_elements(By.CLASS_NAME, "x78zum5")
elements = browser.find_elements(By.CSS_SELECTOR,"div.x1r8uery.x78zum5")
```

```python
import pandas as pd

# Initialize lists to store data
prices_list = []
locations_list = []
car_types_list = []
mileage_list = []
dealership_flag_list = []
price_decrease_flag_list = []
sold_flag_list = []

i =0
for ele in elements:
    if  len(ele.find_elements(By.TAG_NAME, 'img'))==1 and 'miles' in ele.text : 
        #display(HTML(ele.get_attribute('outerHTML')))
        # Extract information
        sold = "N"
        price = ele.find_elements(By.CSS_SELECTOR, "span.xzsf02u.x1943h6x")[0].text
        location = ele.find_elements(By.CSS_SELECTOR, "span.xlyipyv.xuxw1ft")[0].text
        car_type = ele.find_elements(By.CSS_SELECTOR, "div.x1gslohp.xkh6y0r")[1].text
        mileage = ele.find_elements(By.CSS_SELECTOR, "span.xlyipyv.xuxw1ft")[1].text
        dealership_flag = 'Dealership' if 'Dealership' in mileage else 'Private'
        
        mileage = mileage.replace(' · Dealership','')
        price_decrease = "N"
        
        prices = ele.find_elements(By.CSS_SELECTOR, 'div.x78zum5 > span')
        
        print(prices[0].text.upper().strip())
        if prices[0].text.upper().strip() == 'SOLD':
            sold = "Y"
            price = prices[2].text.upper().strip()
        
        if len(prices) > 1: price_decrease = "Y"
        print ("below are the price and price history")
        for p in prices:
            print(p.text)
        i=i+1  
        print('--------')
                
        # Append data to lists
        prices_list.append(price)
        locations_list.append(location)
        car_types_list.append(car_type)
        mileage_list.append(mileage)
        dealership_flag_list.append(dealership_flag)
        price_decrease_flag_list.append(price_decrease)
        sold_flag_list.append(sold)
        
     
    if i >200:
        break

# Create a DataFrame
df = pd.DataFrame({
    'Price': prices_list,
    'Location': locations_list,
    'Car Type': car_types_list,
    'Mileage': mileage_list,
    'Dealership Flag': dealership_flag_list,
    'Price Decrease Flag': price_decrease_flag_list,
    'Sold Flag': sold_flag_list
})

# Print the DataFrame
# Save the DataFrame to a CSV file
df.to_csv('output.csv', index=False)
```
